# Visualizing The Gender Gap In College Degrees

This projet was a completed as a part of the Data Analyst Dataquest series on 07/21/19

**Goal:** Our aim in this project was to analyze the gender gap in S.T.E.M. (Science, Technology, Engineering, and Mathematics) fields compared to other majors. For this we used a dataset compiled by Randal Olson, a data scientist at University of Pennsylvania. He has cleaned the data set made publically available by the Department of Education Statistics. They release a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category, and Randall made it publically available on his [personal website](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/). We will first take a look at just the S.T.E.M. degrees.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Read in the csv file as a pandas dataframe
women_degrees = pd.read_csv('/Users/pgladkevich/Desktop/coding/projects/datasets/percent-bachelors-degrees-women-usa.csv')

# Setting line color values to those standardized to work for color blind people
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# Degrees that are related to the S.T.E.M. field as a list
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

# Change the figure object's size
fig = plt.figure(figsize=(18, 3))

# Iterate and create plots tracking gender percentage of a major category over time
for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

The S.T.E.M. degrees are sorted in order of decreasing gender gap from left to right. The percentage of degree holders that are male vs female differs most greatly in Engineering, while the gap is smallest in Math and Statistics. There are more male degree holders across all degrees except for Psychology and Biology, who flipped from have a majority of male to female in approximately 1970 and 1990.

# Comparing across all degrees



```python
# First group the rest of the degrees into categories
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16,20))


# Recreate the graphs from above but now in a single column
for sp in range(0,18,3):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[int(sp/3)])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
        ax.tick_params(labelbottom=True)
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom=True)
    
for sp in range(1,16,3):
        ax = fig.add_subplot(6,3,sp+1)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[int((sp-1)/3)]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[int((sp-1)/3)]], c=cb_orange, label='Men', linewidth=3)
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(lib_arts_cats[int((sp-1)/3)])
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
        if sp == 1:
            ax.text(2003, 78, 'Women')
            ax.text(2005, 18, 'Men')
            ax.tick_params(labelbottom=True)
        
for sp in range(2,18,3):
        ax = fig.add_subplot(6,3,sp+1)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[int((sp-2)/3)]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[int((sp-2)/3)]], c=cb_orange, label='Men', linewidth=3)
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(other_cats[int((sp-2)/3)])
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
        if sp == 2:
            ax.text(2003, 90, 'Women')
            ax.text(2005, 5, 'Men')
            ax.tick_params(labelbottom=True)
        elif sp == 17:
            ax.text(2005, 62, 'Men')
            ax.text(2003, 30, 'Women')
            ax.tick_params(labelbottom=True)

# Export image for use in a presentation, email, or documentation
plt.savefig("gender_degrees.png")
fig.subplots_adjust(hspace=.3,wspace=.25)
```

# Conclusion

In the three seperate major categories we can easily tell that Engineering and Computer Science have the lowest percentages of degrees given to women; however, the other categories are a little more favorable for women with there even being a majority of female degrees awarded in Biology and especially Psychology. There is also a noticeable difference between degrees awarded in Architecture. It appears that there needs to be a push to encourage more women to pursue these three degrees that are problem categories.
